```python
plt.figure(figsize=(12,7))
sns.boxplot(x = "Pclass", y="Age" , data =train ,palette= "winter")
```

```python
def impute_age(cols):
  Age = cols[0]
  Pclass = cols[1]

  if pd.isnull(Age):
    if Pclass == 1:
      return 37

    elif Pclass == 2:
      return 27

    else:
      return 24
  else:
    return Age
```

```python
train["Age"] = train[["Age", "Pclass"]].apply(impute_age, axis=1)
```

```python
#Analyze the number of passengers who embarked from each port (Cherbourg, Queenstown,
# Southampton) and survival rates for each group.
# Group by port of embarkation and calculate survival rate
grouped_df = df.groupby('Embarked')['Survived'].mean() * 100


# Create a bar chart
grouped_df.plot(kind='bar')

# Add title and labels
plt.title('Survival Rate by Port of Embarkation')
plt.xlabel('Embarked')
plt.ylabel('Survival Rate (%)')

# Show the plot
plt.show()

# Print the number of passengers who embarked from each port
print(df['Embarked'].value_counts())

```

```python
# Calculate the percentage of passengers who embarked from each port
percentages = (df['Embarked'].value_counts() / len(df)) * 100

# Print the results
print(percentages)

plt.pie(percentages,labels =percentages.index, autopct="%1.1f%%")

plt.title("Embarked_Percentage for Passnger of Titanic_train")

plt.show()
```
