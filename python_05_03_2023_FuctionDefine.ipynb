```python
def person_details(first_name, last_name, age = None, country = "United States"):   # function definition
    person = {}  # empty dictionary
    person["first_name"] = first_name
    person["last_name"] = last_name
    if age :
        person["age"] = age
    person["Country"] = country
    return person
```

```python
#ceating function it save the data and return the saved data everytime it runs

def data_store():
  data = {}

  def save (key, value):
     #saving data in key value pair means in dictionary

    data[key] =value

  def share(): #this return the data dictionary which save
    return data.copy()

  save("name", "ARchana Harindra")
  save("Age", 100)
  save("Country", "India")

  return share

share_data = data_store()

print(share_data)

```

```python
#for input received

def data_store():
  data = {}

  def save(data_dict):#saving data in dictionary
     data.update(data_dict)

  def share():
    return data.copy()

  return save, share

save, share= data_store()
user_input = input(str("Enter your full name,Age,  city and Country seperated by commas : "))
user_data = {
    "name" : user_input.split()[0],
    "age" : user_input.split()[1],
    "city": user_input.split()[2],
    "country":user_input.split()[3]
}

save(user_data)
share_data = share()
print(share_data)
```
